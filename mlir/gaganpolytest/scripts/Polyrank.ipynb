### Create Ranking based on GFLOPS

```python
def create_rankings(train_df,index):
    Ranking = []
    Ranking.append(1)
    max_temp = train_df.iloc[0, index]
    Rank =1
    for i in range(1,len(train_df)):
        if train_df.iloc[i, index] == max_temp:
            Ranking.append(Rank)
        else:
            Rank +=1
            max_temp = train_df.iloc[i, index]
            Ranking.append(Rank)
    return Ranking
```

```python
file = '../results/perf.csv'
train_df = pd.read_csv(file,header=None)
print(train_df)

proxyDataAccessTime_list = []
for i in range(len(train_df)) : 
    proxyDataAccessTime_list.append(train_df.iloc[i, 0]*4+train_df.iloc[i, 1]*14+train_df.iloc[i, 2]*60+train_df.iloc[i, 3]*84) 

proxyDataAccessTime_list = pd.DataFrame(proxyDataAccessTime_list) 
train_df = pd.concat([train_df, proxyDataAccessTime_list], axis=1,ignore_index =True)

# idx = 4
for idx in range(4,6):
    train_df = train_df.sort_values(by=[idx],ascending=False)
    if idx == 5:
        train_df = train_df.sort_values(by=[idx],ascending=True)
    train_df = train_df.reset_index(drop=True) 
    rank_order = create_rankings(train_df,idx)
    rank_order = pd.DataFrame(rank_order)
    train_df = pd.concat([train_df, rank_order], axis=1,ignore_index =True)
```
